```python
import pandas as pd
import re
```

```python
df_games = pd.read_json('./Data/liga brasil/games.json')
df_games.head()
```

```python
df_games['events'][0]
```

```python
# Crear un DataFrame vacío para almacenar todos los eventos
df_events = pd.DataFrame()

# Iterar sobre las listas de eventos y otra columna en df_games
for index, row in df_games.iterrows():
    list_events = row['events']
    game_id = row['game_id']
    
    # Convertir la lista de eventos en un DataFrame
    df = pd.DataFrame(list_events)
    
    # Agregar la otra columna a df
    df['game_id'] = game_id
    
    # Concatenar el DataFrame de eventos actual con el DataFrame general de eventos
    df_events = pd.concat([df_events, df], ignore_index=True)

df_events.head()
```

```python
df_events.to_csv('./Data/liga brasil/events.csv',index=False)
```

```python
df_games.drop('events',axis=1,inplace=True)
```

```python
df_games.head()
```

Se eliminan las columnas con valores redundantes

```python
df_games.drop('parent',axis=1,inplace=True)
df_games.drop('type',axis=1,inplace=True)
df_games.drop('href',axis=1,inplace=True)
df_games.head()
```

```python
# Función para extraer el número del diccionario
def extrac_id(diccionario):
    match = re.search(r'/(\d+)/', diccionario['href'])
    if match:
        return int(match.group(1))
    else:
        return None

# Aplicar la función a cada fila de la columna
df_games['home_club_id'] = df_games['home_club'].apply(extrac_id)
df_games['away_club_id'] = df_games['away_club'].apply(extrac_id)
df_games.drop('home_club',axis=1,inplace=True)
df_games.drop('away_club',axis=1,inplace=True)
df_games.head()
```

Se normalizan columnas y se transforman valores

```python
df_games['home_club_position'] = df_games['home_club_position'].str.replace('Position: ', '').astype(int)
df_games['away_club_position'] = df_games['away_club_position'].str.replace('Position: ', '').astype(int)
df_games['matchday'] = df_games['matchday'].str.replace('. Matchday', '').astype(int)
df_games['attendance'] = df_games['attendance'].str.replace('Attendance: ','').str.replace('.', '').astype(int)
df_games.head()
```

```python
df_games['home_manager'] = df_games['home_manager'].apply(lambda x: x['name'] if isinstance(x, dict) and 'name' in x else None)
df_games.head()
```

```python
df_games['away_manager'] = df_games['away_manager'].apply(lambda x: x['name'] if isinstance(x, dict) and 'name' in x else None)
df_games.head()
```

```python
# Función para transformar la cadena en una tupla de enteros
def transformar_a_tupla(cadena):
    partes = cadena.split(":")
    return (int(partes[0]), int(partes[1]))

# Aplicar la función a cada fila de la columna 'resultado'
df_games['result'] = df_games['result'].apply(transformar_a_tupla)
df_games.head()
```

```python
df_games.to_csv('./Data/liga brasil/games.csv',index=False)
```
